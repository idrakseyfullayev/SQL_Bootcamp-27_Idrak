## SQL-də **implicit (dolayı)** və **explicit (açıq)** çevrilmə anlayışları, **data type conversion** (məlumat tipinin dəyişdirilməsi) ilə bağlıdır:




### 🔸 **Implicit Conversion** (Dolayı çevrilmə)

* SQL özü avtomatik məlumat tipini uyğunlaşdırır.
* İstifadəçi çevrilmə əmri yazmır.

**Misal:**

```sql
SELECT * FROM employees WHERE salary = '1000';
```

Burada `salary` ədədi tipdədir, `'1000'` isə stringdir. SQL avtomatik `'1000'` → `1000` çevirir.

✅ **Avtomatik baş verir**

---

### 🔸 **Explicit Conversion** (Açıq çevrilmə)

* İstifadəçi çevrilməni açıq şəkildə göstərəndə baş verir.

**Misal:**

```sql
SELECT TO_NUMBER('1000') FROM dual;
```

və ya

```sql
SELECT TO_CHAR(hire_date, 'YYYY-MM-DD') FROM employees;
```

✅ **İstifadəçi özü `TO_CHAR()`, `TO_DATE()`, `CAST()` və s. ilə yazır**

---

### Nəticə:

| Növ      | Nümunə              | Kim edir?  |
| -------- | ------------------- | ---------- |
| Implicit | `salary = '1000'`   | SQL özü    |
| Explicit | `TO_NUMBER('1000')` | İstifadəçi |

Sadə dillə desək:

* **Implicit** → SQL özü çevirir
* **Explicit** → sən ona "çevir" deyirsən

İstəsən real data ilə izah da verə bilərəm.

### AGGREGATE FUNCTIONS

- **Diqqət etmək lazımıdr ki bu Funksiyaları tətəbiq edərkən əgər NULL dəyərlər olarsa Nəticə çox vaxt NULL olacaq**
    - **Çıxış yolu: `WHERE salary IS NOT NULL;` - NULL dəyəri filter ilə aradan çıxardırıq**

**`MIN(salary)`** → maaşların **ən kiçik dəyərini** qaytarır. **Nümunə:** `MIN(salary) FROM hr.employees` → `3000`

**`MAX(salary)`** → maaşların **ən böyük dəyərini** qaytarır. **Nümunə:** `MAX(salary) FROM hr.employees` → `24000`

**`SUM(salary)`** → maaşların **cəmini** hesablayır. **Nümunə:** `SUM(salary) FROM hr.employees` → `450000`

**`COUNT(*)`** → **sətir sayını** hesablayır. **Nümunə:** `COUNT(*) FROM hr.employees` → `107`

**`AVG(salary)`** → maaşların **orta dəyərini** qaytarır. **Nümunə:** `AVG(salary) FROM hr.employees` → `4205.61`

<br>

**`GROUP BY department_id`** → verilənləri **department_id üzrə qruplaşdırır**, **hər qrup üçün ayrıca hesablama** etməyə imkan verir. 

**Nümunə:**

```sql
SELECT department_id, AVG(salary) AS orta_maas
FROM hr.employees
GROUP BY department_id;
```
→ Hər department_id üçün orta maaşı qaytarır.

<br>

**`HAVING`** → `GROUP BY` ilə yaradılmış qruplara **şərt tətbiq etmək** üçün istifadə olunur.
- `WHERE` ilə fərqi: `WHERE` **sətirlərə**, `HAVING` isə **qruplara** şərt qoyur.

**Nümunə:**

```sql
SELECT department_id, AVG(salary) AS orta_maas
FROM hr.employees
GROUP BY department_id
HAVING AVG(salary) > 5000;
```
→ Yalnız **orta maaşı 5000-dən böyük olan** department_id-ləri göstərir.





### JOINS

**`INNER JOIN`** → iki cədvəli **ortaq sütun üzrə birləşdirir** və yalnız **hər iki cədvəldə uyğun gələn sətirləri** qaytarır.

**Nümunə:**

```sql
SELECT e.employee_id, e.first_name, d.department_name
FROM hr.employees e
INNER JOIN hr.departments d
ON e.department_id = d.department_id;
```
→ Yalnız `employees` və `departments` cədvəllərində **department_id uyğun gələn** sətirlər çıxacaq.

<br>

**`LEFT JOIN`** → iki cədvəli **sol tərəfdəki cədvəl əsas olmaqla birləşdirir** və **sol cədvəldəki bütün sətirləri** qaytarır, sağ cədvəldə uyğun sətir yoxdursa, həmin sətirlərə `NULL` qoyur.

**Nümunə:**

```sql
SELECT e.employee_id, e.first_name, d.department_name
FROM hr.employees e
LEFT JOIN hr.departments d
ON e.department_id = d.department_id;
```
→ Bütün `employees` sətirləri çıxacaq, amma bəzi `department_id`-lər `departments`-də yoxdursa, `department_name` sütununun həmin sətirləri `NULL` olacaq.

<br>

**`RIGHT JOIN`** → iki cədvəli **sağ tərəfdəki cədvəl əsas olmaqla birləşdirir** və **sağ cədvəldəki bütün sətirləri** qaytarır, sol cədvəldə uyğun sətir yoxdursa, həmin sətirlərə `NULL` qoyur.

**Nümunə:**

```sql
SELECT e.employee_id, e.first_name, d.department_name
FROM hr.employees e
RIGHT JOIN hr.departments d
ON e.department_id = d.department_id;
```
→ Bütün `departments` sətirləri çıxacaq, amma bəzi `employee`-lər uyğun gəlmirsə, `employee_id` və `first_name` sütununun həmin sətirləri `NULL` olacaq.

<br>

**`FULL JOIN`** → iki cədvəli birləşdirir və **hər iki cədvəldəki bütün sətirləri** qaytarır. Uyğun sətir yoxdursa, həmin tərəfin sütunlarının həmin sətirlərinə `NULL` qoyur.

**Nümunə:**

```sql
SELECT e.employee_id, e.first_name, d.department_name
FROM hr.employees e
FULL JOIN hr.departments d
ON e.department_id = d.department_id;
```
→ Hər iki cədvəldəki bütün sətirlər çıxacaq. Uyğun sətir yoxdursa, uyğun olmayan sətirlər `NULL` olacaq.

<br>

**`CROSS JOIN`** → iki cədvəli birləşdirir və **hər bir sətirdən digər cədvəldəki bütün sətirlərlə kombinasiya** yaradır. Nəticədə **Kartesian hasil** (cross product) əmələ gəlir.

**Nümunə:**

```sql
SELECT e.employee_id, e.first_name, d.department_name
FROM hr.employees e
CROSS JOIN hr.departments d;
```
→ `employees`-dəki hər sətir `departments`-dəki hər sətirlə birləşəcək, nəticədə çox sayda sətir yaranacaq.

<br>

**`SELF JOIN`** → **eyni cədvəl üzərində öz-özünə birləşmə**dir. Adətən, cədvəldəki sətirlər arasında nisbət və ya əlaqə tapmaq üçün istifadə olunur. `AS` ilə cədvəli fərqli adlarla təmsil edirik.

**Nümunə:**

```sql
SELECT e1.employee_id AS işçi, e1.manager_id, e2.first_name AS menecer_adı
FROM hr.employees e1
LEFT JOIN hr.employees e2
ON e1.manager_id = e2.employee_id;
```
→ Hər işçinin **menecerini eyni cədvəldən** tapır.

- **İzah:**

> `SELECT FIRST_NAME, EMPLOYEE_ID, MANAGER_ID FROM hr.employees` **bu kodu işlədikdə görərik ki, hər işçinin qabağındakı `MANAGER_ID` onun menecerinin ID-sidir və həmin`MANAGER_ID` öz növbəsində hemin menecrin öz `EMPLOYEE_ID`-sidir.**

<br>

`ON emp.department_id = dep.department_id AND emp.salary > 5000` Əvvəlcə **employees** cədvəlində **maaşı 5000-dən böyük olan işçiləri** filter edir, sonra **employees və departments cədvəllərini birləşdirir**:

```sql
SELECT emp.first_name, emp.last_name, emp.salary, 
       dep.department_name, dep.department_id
FROM hr.employees emp
JOIN hr.departments dep 
ON emp.department_id = dep.department_id AND emp.salary > 5000;
```

🔹 **İzahı:**
- **Əvvəlcə işçilər cədvəli filtr edilir, sonra birləşdirilir**
- **`AND emp.salary > 5000` → birləşmə zamanı yalnız maaşı 5000-dən böyük işçilər nəzərə alınır.**

### **SET OPERATORS**

**`UNION ALL`** → iki və ya daha çox sorğunun nəticələrini **alt-alta birləşdirir** və **bütün sətirləri (təkrarlananlar daxil) saxlayır**.

**Nümunə:**

```sql
SELECT country_name
FROM hr.countries
WHERE region_id = 10
UNION ALL
SELECT country_name
FROM hr.countries
WHERE region_id = 10 OR region_id = 20
ORDER BY country_name;
```
→ Nəticələr **alt-alta gəlir** və təkrar dəyərlər silinmir.

<br>

**`UNION`** → iki və ya daha çox sorğunun nəticələrini **alt-alta birləşdirir** və **təkrarlanan sətirləri silir**.

**Nümunə:**

```sql
SELECT country_name
FROM hr.countries
WHERE region_id = 10
UNION
SELECT country_name
FROM hr.countries
WHERE region_id = 10 OR region_id = 20
ORDER BY country_name;
```
→ Nəticələr **alt-alta gəlir**, amma **duplicate (təkrar) dəyərlər çıxarılır**.

<br>

**`INTERSECT`** → iki sorğunun nəticələrindən **yalnız ortaq (kəsişən) sətirləri** qaytarır.

**Nümunə:**

```sql
SELECT country_name
FROM hr.countries
WHERE region_id = 10
INTERSECT
SELECT country_name
FROM hr.countries
WHERE region_id = 10 OR region_id = 20
ORDER BY country_name;
```
→ Hər iki sorğuda **eyni olan dəyərlər** nəticədə göstərilir.

<br>

**`MINUS`** → **birinci sorğunun** nəticəsindən **ikinci sorğuda olanları çıxır**, yəni yalnız **birincidə olub ikincidə olmayan sətirləri** qaytarır.

**Nümunə:**

```sql
SELECT country_name
FROM hr.countries
WHERE region_id = 10 OR region_id = 20
MINUS
SELECT country_name
FROM hr.countries
WHERE region_id = 10 
ORDER BY country_name;
```
→ **Birinci sorğuda olub, ikinci sorğuda olmayan dəyərlər göstərilir.**


**`LAG(salary, 1, 0) OVER (ORDER BY hire_date)`** → **özündən əvvəlki sətirin** (**1 əvvəlki**) `salary` dəyərini qaytarır, əgər **əvvəlki sətir yoxdursa** **0 verir**.

**Nümunə:**
```sql
SELECT last_name, hire_date, salary,
LAG(salary, 1, 0) OVER (ORDER BY hire_date)
FROM hr.employees 
WHERE job_id = 'PU_CLERK';
```
→ **özündən əvvəlki sətirdə olan işçinin maaşını göstərir, ilk sətir üçün `0` olur.**

<br>

**`LEAD(salary, 1, 0) OVER (ORDER BY hire_date)`** → növbəti sətirin (**1 sonrakı**) `salary` dəyərini qaytarır, əgər növbəti sətir yoxdursa **0 verir**.

**Nümunə:** 
```sql
SELECT hire_date, last_name, salary,
LEAD(salary, 1) OVER (ORDER BY hire_date) AS "NextHired"
FROM hr.employees
WHERE department_id = 30;
```
→ **növbəti sətirdəki işçinin maaşını göstərir, son sətir üçün `0` olur.**

<br>

**`LAST_VALUE(salary) OVER (ORDER BY EMPLOYEE_ID ROWS BETWEEN CURRENT ROW AND 3 FOLLOWING)`** → **cari sətirdən başlayaraq növbəti 3 sətir daxil olmaqla pəncərədəki son maaşı** qaytarır.

**Nümunə:**

```sql id="1l2k5q"
SELECT last_name, salary,
       LAST_VALUE(salary) OVER (
           ORDER BY EMPLOYEE_ID
           ROWS BETWEEN CURRENT ROW AND 3 FOLLOWING
       ) AS last_val
FROM hr.employees;
```

🔹 **İzahı:**

* `ORDER BY EMPLOYEE_ID` → sətirləri `EMPLOYEE_ID`-yə görə düzür.
* `ROWS BETWEEN CURRENT ROW AND 3 FOLLOWING` → **cari sətirdən başlayaraq növbəti 3 sətir pəncərədir**.
* `LAST_VALUE(salary)` → bu pəncərədəki **ən son (sonuncu) salary** göstərilir.

**Qısa başa düşmək:**

* Hər sətir üçün **cari sətirdən 3 aşağıya qədər baxılır**
* Baxılan sətirlərdən **ən son salary** həmin sətir üçün çıxır.

<br>

**`FIRST_VALUE(salary) OVER (ORDER BY employee_id RANGE BETWEEN 3 PRECEDING AND CURRENT ROW)`** → cari sətir və **onun əvvəlki 3 sətirinə** baxır və bu pəncərədəki **ilk `salary` dəyərini** qaytarır.

**Nümunə:**

```sql id="2h3k5q"
SELECT last_name, salary,
       FIRST_VALUE(salary) OVER (
           ORDER BY employee_id 
           RANGE BETWEEN 3 PRECEDING AND CURRENT ROW
       ) AS first_val
FROM hr.employees;
```

🔹 **İzahı:**

* `ORDER BY employee_id` → sətirləri `employee_id`-yə görə düzür.
* `RANGE BETWEEN 3 PRECEDING AND CURRENT ROW` → cari sətir və ondan əvvəlki **3 employee_id dəyərini** pəncərə kimi götürür.
* `FIRST_VALUE(salary)` → bu pəncərədə **ən əvvəlki maaşı** göstərir.

💡 **Qısa başa düşmək:**

* Hər sətir üçün **pəncərə cari + əvvəlki 3 sətirdir**.
* Pəncərədəki **ilk salary** həmin sətir üçün `first_val` olur.
