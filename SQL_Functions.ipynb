### Excel  →  Oracle SQL Text Functions – Qısa izahlar:

**LEFT** → `SUBSTR()`
Soldan simvol çıxarır → `SUBSTR('Excel', 1, 2)` → **Ex**

**RIGHT** → `SUBSTR()`
Sağdan simvol çıxarır → `SUBSTR('Excel', -2)` → **el**

**MID** → `SUBSTR()`
Ortadan simvol çıxarır → `SUBSTR('Excel', 2, 2)` → **xc**

**LEN** → `LENGTH()`
Simvol sayını verir → `LENGTH('Excel')` → **5**

**REPT** → `RPAD()` və ya `LPAD()`
Mətni təkrar edir → `RPAD('X', 3, 'X')` → **XXX**

**CONCATENATE** → `CONCAT()` *(iki dəyər üçün)*
Mətin birləşdirir → `CONCAT('Ex','cel')` → **Excel**

**CONCAT** → `'||'` *(çoxlu dəyər üçün)*
Mətin birləşdirir → `'Ex' || 'cel'` → **Excel**

**TEXTJOIN** → `LISTAGG()`
Ayırıcı ilə birləşdirir → `LISTAGG(column_name, '-') WITHIN GROUP (ORDER BY id)` → **A-B-C**

**TEXTJOIN (range)** → `LISTAGG()` + `DISTINCT`, daha kompleks hallarda `XMLAGG`
Range üçün tətbiq olunur → `SELECT LISTAGG(col, ' ') WITHIN GROUP (...)` → **A B, C D, ...**

**EXACT** → `= (və ya CASE)`
Case-sensitive müqayisə üçün → `A = B COLLATE binary_cs` və ya uyğun `COLLATE` ilə

**FIND** → `INSTR()`
Case-sensitive axtarır → `INSTR('Excel', 'c')` → **3**

**SEARCH** → `INSTR(UPPER(...))`
Case-insensitive → `INSTR(UPPER('Excel'), 'C')` → **3**

**LEFT+FIND** → `SUBSTR(STRING, 1, INSTR(STRING, ' ') - 1)`
Birinci sözü verir

**LOWER** → `LOWER()`
Mətni kiçik hərfə çevirir → `LOWER('Excel')` → **excel**

**UPPER** → `UPPER()`
Mətni böyük hərfə çevirir → `UPPER('Excel')` → **EXCEL**

**PROPER** → `INITCAP()`
Hər sözün ilk hərfini böyük edir → `INITCAP('excel is fun')` → **Excel Is Fun**

**FIND+LOWER** → `INSTR(LOWER('Excel'), 'e')`
Kiçik hərflə simvol yeri → **1**

**REPLACE** → Oracle-da `STUFF()` funksiyası yoxdur
Mövqeyə əsasən dəyişdirmə üçün `SUBSTR()` və `||` ilə qurulur

**SUBSTITUTE** → `REPLACE()`
Mətni hissə ilə dəyişir → `REPLACE('Excel','cel','tend')` → **Extentd**

**TRIM** → `TRIM()`
Artıq boşluqları silir → `TRIM('  Excel  ')` → **Excel**

**TEXT** → `TO_CHAR()`
Formatlı mətnə çevirir → `TO_CHAR(1234.5, '$999,999.00')` → **\$1,234.50**

**VALUE** → `TO_NUMBER()`
Mətn tipində olan dəyəri ədədi dəyərə çevirir → `TO_NUMBER('123.45')` → **123.45**

\**--, *1, +0, /1** → `TO_NUMBER()` və ya sadəcə `*1`
Mətn tipində olan dəyəri ədədi dəyərə çevirmək üçün → `TO_NUMBER('1234')` və ya `'1234'*1`

**FIXED** → `TO_CHAR(ROUND(...))`
Ədədi müəyyən onluq yerə yuvarlaqlaşdırır və mətn kimi qaytarır → `TO_CHAR(ROUND(123.4567, 2), '999.99')` → **123.46**

## SQL-də **implicit (dolayı)** və **explicit (açıq)** çevrilmə anlayışları, **data type conversion** (məlumat tipinin dəyişdirilməsi) ilə bağlıdır:




### 🔸 **Implicit Conversion** (Dolayı çevrilmə)

* SQL özü avtomatik məlumat tipini uyğunlaşdırır.
* İstifadəçi çevrilmə əmri yazmır.

**Misal:**

```sql
SELECT * FROM employees WHERE salary = '1000';
```

Burada `salary` ədədi tipdədir, `'1000'` isə stringdir. SQL avtomatik `'1000'` → `1000` çevirir.

✅ **Avtomatik baş verir**

---

### 🔸 **Explicit Conversion** (Açıq çevrilmə)

* İstifadəçi çevrilməni açıq şəkildə göstərəndə baş verir.

**Misal:**

```sql
SELECT TO_NUMBER('1000') FROM dual;
```

və ya

```sql
SELECT TO_CHAR(hire_date, 'YYYY-MM-DD') FROM employees;
```

✅ **İstifadəçi özü `TO_CHAR()`, `TO_DATE()`, `CAST()` və s. ilə yazır**

---

### Nəticə:

| Növ      | Nümunə              | Kim edir?  |
| -------- | ------------------- | ---------- |
| Implicit | `salary = '1000'`   | SQL özü    |
| Explicit | `TO_NUMBER('1000')` | İstifadəçi |

Sadə dillə desək:

* **Implicit** → SQL özü çevirir
* **Explicit** → sən ona "çevir" deyirsən

İstəsən real data ilə izah da verə bilərəm.




### **Excel  →  Oracle SQL Date Functions – Qısa izahlar:**

**DATE** → `TO_DATE()`
Tarix yaradır → `TO_DATE('2025-04-25', 'YYYY-MM-DD')` → **25-APR-25**

**DAY** → `TO_CHAR(..., 'DD')`
Günü çıxarır → `TO_CHAR(DATE '2025-04-25', 'DD')` → **25**

**MONTH** → `TO_CHAR(..., 'MM')`
Ayı çıxarır → `TO_CHAR(DATE '2025-04-25', 'MM')` → **04**

**YEAR** → `TO_CHAR(..., 'YYYY')`
İli çıxarır → `TO_CHAR(DATE '2025-04-25', 'YYYY')` → **2025**

**DATEVALUE** → `TO_DATE()`
Mətni tarixə çevirir → `TO_DATE('2025-04-25', 'YYYY-MM-DD')` → **25-APR-25**

**NOW** → `SYSDATE`
Cari tarix və vaxt → `SYSDATE` → **25-APR-25 10:15:00**

**TODAY** → `TRUNC(SYSDATE)`
Cari tarix (vaxtsız) → `TRUNC(SYSDATE)` → **25-APR-25**

**HOUR** → `TO_CHAR(..., 'HH24')`
Saatı çıxarır → `TO_CHAR(SYSDATE, 'HH24')` → **10**

**MINUTE** → `TO_CHAR(..., 'MI')`
Dəqiqəni çıxarır → `TO_CHAR(SYSDATE, 'MI')` → **15**

**SECOND** → `TO_CHAR(..., 'SS')`
Saniyəni çıxarır → `TO_CHAR(SYSDATE, 'SS')` → **30**

**TIME** → `TO_DATE(..., 'HH24:MI:SS')`
Vaxt yaradır → `TO_DATE('10:15:30', 'HH24:MI:SS')` → **10:15:30**

**TIMEVALUE** → `TO_NUMBER(TO_CHAR(..., 'SSSSS'))/86400`
Vaxtı ədədə çevirir → `(TO_NUMBER(TO_CHAR(DATE '2025-04-25 10:15:30', 'SSSSS')) / 86400)` → **0.427083**

**EDATE** → `ADD_MONTHS()`
Tarixi ay üzrə dəyişir → `ADD_MONTHS(DATE '2025-04-25', 2)` → **25-JUN-25**

**EOMONTH** → `LAST_DAY(ADD_MONTHS(...))`
Ayın son gününü qaytarır → `LAST_DAY(ADD_MONTHS(DATE '2025-04-25', 1))` → **31-MAY-25**

**NETWORKDAYS** → `NEXT_DAY()`, `HOLIDAY COUNT`
İş günləri fərqi → PL/SQL və ya xüsusi funksiyalarla sayılır

**WORKDAY** → `NEXT_DAY()` + `n iş günü`
İş günü sonrası tarix → PL/SQL və ya loop ilə qurulur

**DAYS** → `DATE1 - DATE2`
İki tarix arasındakı gün fərqi → `DATE '2025-04-25' - DATE '2025-04-20'` → **5**

**WEEKDAY** → `TO_CHAR(..., 'D')`
Həftənin günü (1 = Bazar) → `TO_CHAR(DATE '2025-04-25', 'D')` → **6 (Cümə)**

**WEEKNUM** → `TO_CHAR(..., 'WW')`
İlin neçəci həftəsidir → `TO_CHAR(DATE '2025-04-25', 'WW')` → **17**

**WEEKNUM (b.e. ilə)** → `TO_CHAR(..., 'IW')`
ISO həftə nömrəsi (bazar ertəsi ilə) → `TO_CHAR(DATE '2025-04-25', 'IW')` → **17**

**YEARFRAC** → `(DATE2 - DATE1)/365`
İki tarix arasındakı il fərqi → `(DATE '2025-04-25' - DATE '2025-01-01') / 365` → **0.31**

**DATEDIF** → `MONTHS_BETWEEN`, `TRUNC`, `FLOOR`
İllik, aylıq, gündəlik fərqlər:

* **İl fərqi** → `TRUNC(MONTHS_BETWEEN(D2,D1)/12)`
* **Ay fərqi** → `TRUNC(MONTHS_BETWEEN(D2,D1))`
* **Gün fərqi** → `D2 - D1`
* **YM** → `MOD(TRUNC(MONTHS_BETWEEN(D2,D1)), 12)`
* **MD** və **YD** → PL/SQL ilə qurulur
